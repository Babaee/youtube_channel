```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from matplotlib import animation
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import PillowWriter
```

$\vec{S} = (x_1, y_1, x_2, y_2, ...)$

```python
G = 1
m1 = 1
m2 = 2 #333000 (sun)
x1_0 = 1
y1_0 = 0
x2_0 = 0
y2_0 = 0
vx1_0 = 0
vy1_0 = np.sqrt(m2) #v=sqrt(a*r**2) with a=m2 (since G=1, r12=1)
vx2_0 = 0
vy2_0 = 0
```

```python
def dSdt(S, t):
    x1, y1, x2, y2, vx1, vy1, vx2, vy2 = S
    r12 = np.sqrt((x2-x1)**2 + (y2-y1)**2)
    return [ vx1,
            vy1,
            vx2,
            vy2,
            m2/r12**3 * (x2-x1),
            m2/r12**3 * (y2-y1),
            m1/r12**3 * (x1-x2),
            m1/r12**3 * (y1-y2) ]
```

```python
dSdt([x1_0, y1_0, x2_0, y2_0, vx1_0, vy1_0, vx2_0, vy2_0], 0)
```

```python
# t = np.linspace(0,1,10000) #works for sun
t = np.linspace(0,1000,10000)
```

```python
sol = odeint(dSdt, y0=[x1_0, y1_0, x2_0, y2_0, vx1_0, vy1_0, vx2_0, vy2_0],
             t=t)
```

```python
plt.plot(sol.T[0])
plt.xlim(0,200)
plt.grid()
```

```python

tt = 1/np.sqrt(6.67e-11 * 5.97e24 / (1.5e11)**3 ) # seconds
tt = tt / (60*60 * 24* 365.25) / np.diff(t)[0] # per time step (in years)
```

```python
x1 = sol.T[0]
y1 = sol.T[1]
x2 = sol.T[2]
y2 = sol.T[3]
```

```python
def animate(i):
    ln1.set_data([x1[i], x2[i]], [y1[i], y2[i]])
    text.set_text('Time = {:.2f} Years'.format(i*tt))
fig, ax = plt.subplots(1,1, figsize=(8,8))
ax.grid()
ln1, = plt.plot([], [], 'ro--', lw=3, markersize=8)
text = plt.text(0.7, 0.7, '')
ax.set_ylim(-5, 5)
ax.set_xlim(-5,5)
ani = animation.FuncAnimation(fig, animate, frames=200, interval=50)
ani.save('plan.gif',writer='pillow',fps=30)
```

# 3 Body Problem

```python
G = 1
m1 = 1
m2 = 2 #333000 (sun)
m3 = 3 #333000 (sun)
x1_0 = 1
y1_0 = 0
x2_0 = 0
y2_0 = 0
x3_0 = -1
y3_0 = 0
vx1_0 = -3
vy1_0 = np.sqrt(m2) #v=sqrt(a*r**2) with a=m2 (since G=1, r12=1)
vx2_0 = 0
vy2_0 = 0
vx3_0 = 1
vy3_0 = -np.sqrt(m2)/3
```

```python
def dSdt(S, t):
    x1, y1, x2, y2, x3, y3, vx1, vy1, vx2, vy2, vx3, vy3 = S
    r12 = np.sqrt((x2-x1)**2 + (y2-y1)**2)
    r13 = np.sqrt((x3-x1)**2 + (y3-y1)**2)
    r23 = np.sqrt((x2-x3)**2 + (y2-y3)**2)
    return [ vx1,
            vy1,
            vx2,
            vy2,
            vx3,
            vy3,
            m2/r12**3 * (x2-x1) + m3/r13**3 * (x3-x1), #mass 1
            m2/r12**3 * (y2-y1) + m3/r13**3 * (y3-y1),
            m1/r12**3 * (x1-x2) + m3/r23**3 * (x3-x2), #mass 2
            m1/r12**3 * (y1-y2) + m3/r23**3 * (y3-y2),
            m1/r13**3 * (x1-x3) + m2/r23**3 * (x2-x3),
            m1/r13**3 * (y1-y3) + m2/r23**3 * (y2-y3)#mass 3
           ]
```

```python
t = np.linspace(0,300,10000)
```

```python
sol = odeint(dSdt, y0=[x1_0, y1_0, x2_0, y2_0, x3_0, y3_0,
                       vx1_0, vy1_0, vx2_0, vy2_0, vx3_0, vy3_0],
             t=t)
```

```python
x1 = sol.T[0]
y1 = sol.T[1]
x2 = sol.T[2]
y2 = sol.T[3]
x3 = sol.T[4]
y3 = sol.T[5]
```

```python
def animate(i):
    ln1.set_data([x1[i], x2[i], x3[i]], [y1[i], y2[i], y3[i]])
    text.set_text('Time = {:.2f} Years'.format(i*tt))
fig, ax = plt.subplots(1,1, figsize=(8,8))
ax.grid()
ln1, = plt.plot([], [], 'ro', lw=3, markersize=8)
text = plt.text(0.7, 0.7, '')
ax.set_ylim(-5, 5)
ax.set_xlim(-5,5)
ani = animation.FuncAnimation(fig, animate, frames=200, interval=50)
ani.save('plan.gif',writer='pillow',fps=30)
```
